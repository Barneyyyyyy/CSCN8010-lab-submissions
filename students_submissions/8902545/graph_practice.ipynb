```python
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy.stats import norm
from collections import Counter
import pandas as pd
from sklearn import datasets
import seaborn as sns
```

```python
range = 5 + 2.5 * np.random.randn(1, 40)
lis = np.sort(range[0])
```

## creating a plt.plot() for a normal distribution

```python
plt.plot(lis ,norm.pdf(lis, 5, 2.5))
```

## boxplot for the iris dataset features

```python
iris = datasets.load_iris()
```

```python
iris_X,iris_y = datasets.load_iris(return_X_y = True)
```

```python
fig = plt.figure(figsize =(10, 7))
# Creating plot
plt.boxplot(iris_X)
# show plot
plt.show()
```

## frequency plot for target classes

```python
count = Counter(iris_y)
df = pd.DataFrame.from_dict(count, orient='index')
df.plot(kind='barh')
```
