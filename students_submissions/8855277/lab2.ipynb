```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

Figure 1: random data by matplotlib

```python
np.random.seed(23456324) # generate random seed
data = {'ran1': np.random.randint(-10, 10, 20),
        'ran2': np.random.randint(-10, 10, 20)}
fig, ax = plt.subplots(figsize = (4, 4))
ax.set_xlabel('random 1')
ax.set_ylabel('random 2')
ax.plot('ran1', 'ran2', data=data)

```
