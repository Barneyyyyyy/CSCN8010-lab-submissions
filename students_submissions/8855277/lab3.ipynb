```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn import datasets
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

```python
X = diabetes_X[:, 2].reshape(-1,1)
# split data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(X, diabetes_y, test_size=0.2, random_state=0)
```

```python
# train the model
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# make predictions using test data
y_pred = model.predict(x_test)
```

```python
# the mean squared error
print('Mean squared error: %.2f' % metrics.mean_squared_error(y_test, y_pred))
```

```python
# plot the test data
plt.scatter(x_test, y_test, color='red')
```

```python
# plot the model
plt.plot(x_test, y_pred, color='blue')
```

```python
# plot the train data distribution
plt.scatter(x_train, y_train, color='red')
```

```python
# plot the model to get an insight of the fitness of the model
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_pred, color='blue')
plt.xlabel("Feature 3: BMI body mass index(scaled)")
plt.ylabel("Target: Diabetes progression one year after baseline")
plt.title("Training: Linear Regression between bmi and diabetes")
plt.show()
```

Report the model parameters

```python
# slope
m = model.coef_[0]
print('Coefficients/Slope: ', m)

# intercept
b = model.intercept_
print('Intercept: ', b)

# Mean Abosolute Error
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error: ', mae)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error: ', mse)

```

```python
# plot the line
plt.scatter(x_test, y_test, color='red')
plt.plot(x_test, y_pred, color='blue')
plt.xlabel("Feature 3: BMI body mass index(scaled)")
plt.ylabel("Target: Diabetes progression one year after baseline")
plt.title("Prediction: predicted target value for a new data point")

# add new data point
plt.scatter(-0.025, m*-0.025+b, color='green')

# add annotation indicating the the new data point
plt.annotate('new data point', xy=(-0.025, m*-0.025+b), xytext=(-0.050, m*-0.050 + b + 100), color='green', arrowprops=dict(facecolor='green'))
plt.show()
```

Take away:

As we can see above, the regression line fits the test data not really well. 

It is not a good model for making predictions.

Using non-linear regression model might fit better.
