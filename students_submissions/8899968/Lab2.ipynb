## Bargraph using Matplotlib 
### This bargraphs illustrate the price percentage increase on electronic Items in Canada from 2022 to 2023

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(5, 6))
electronicItems = ['Mobile Phones', 'Washing Machine', 'Television', 'Gaming Consoles']
price = [30,45,70,60]

ax.bar(electronicItems, price, width = 0.5,color = 'green')
plt.xlabel('Electronic Items')
plt.ylabel('Percentage of Increase')
plt.show()
```

## Pie Charts using matplotlib
### Environmental Polution Representation Using Matplotib Pie Chart

```python
pollutants = ['Vehicles', 'Industries', 'Plastics','Chemical Fertilizers']
percentage = [20,30,60,20 ]
fig, ax = plt.subplots()
ax.pie(percentage,labels=pollutants,autopct = '%1.1f%%',pctdistance=1.25, labeldistance=.6)
plt.show()
```

## Scatter Plot Using Matlpotib

```python
x = np.array([2,3,4,5])
y = np.array([1,3,4,5])

plt.scatter(x, y,color='Green')
plt.xlabel('Random label of X-axis')
plt.ylabel('Random label for Y-axis')
plt.show()
```
