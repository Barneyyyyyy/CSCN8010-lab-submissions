```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

A Bar graph


```python
x = np.linspace(-10, 120, 130)
y = np.sin(x)

plt.plot(x, y)
plt.title('Tan Wave')
plt.xlabel('X Values')
plt.ylabel('Y Values')
plt.show()
```

Line Plot


```python

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.plot(x, y)
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Histogram

```python
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.title("Histogram")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

Bar Plot

```python
labels = ['A', 'B', 'C', 'D']
values = [10, 15, 30, 20]

plt.bar(labels, values)
plt.title("Bar Plot")
plt.show()
```

Box plot

```python
import seaborn as sns

data = np.random.randn(1000)

sns.boxplot(data)
plt.title("Box Plot")
plt.show()
```

Donut Chart

```python
labels = 'A', 'B', 'C', 'D'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0)

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.title("Donut Chart")
plt.show()
```
