```python
from tensorflow.keras.datasets import fashion_mnist
import matplotlib.pyplot as plt

# Loading the fashion MNIST dataset
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

# Exploring the dataset
num_train, img_rows, img_cols = train_images.shape
num_test = test_images.shape[0]
num_classes = len(set(train_labels))

# Displaying the basic information about the dataset
print(f'Number of Training Images: {num_train}')
print(f'Number of Test Images: {num_test}')
print(f'Image Dimensions: {img_rows} x {img_cols} pixels')
print(f'Number of Classes: {num_classes}')

# Displaying some images from the dataset
# First 50 then
plt.figure(figsize=(10, 10))
for i in range(50):
    plt.subplot(10, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(train_images[i], cmap=plt.cm.binary)
    plt.xlabel(train_labels[i])
plt.show()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.optimizers import Adam

# Data Preprocessing
# I normalized the pixel here to get a scale value in range from 0 to 1
train_images = train_images / 255.0
test_images = test_images / 255.0

# Building my Model
model = Sequential([
    Flatten(input_shape=(img_rows, img_cols)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),  
    Dense(num_classes, activation='softmax')
])

# Compiling my Model
model.compile(optimizer=Adam(), 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Training my Model
history = model.fit(train_images, train_labels, epochs=13, validation_split=0.13) 
```

```python
# Building the second model configuration
model_2 = Sequential([
    Flatten(input_shape=(img_rows, img_cols)),
    Dense(256, activation='relu'),  
    Dense(128, activation='relu'),  
    Dense(num_classes, activation='softmax')
])

# Compiling the second model configuration
model_2.compile(optimizer=Adam(),
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

# Training the second model configuration
history_2 = model_2.fit(train_images, train_labels, epochs=15, validation_split=0.15)  
```

```python
# Building the third model configuration
model_3 = Sequential([
    Flatten(input_shape=(img_rows, img_cols)),
    Dense(512, activation='relu'), 
    Dense(num_classes, activation='softmax')
])

# Compiling the third model configuration with a different learning rate
model_3.compile(optimizer=Adam(learning_rate=0.0001),
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])

# Training the third model configuration
history_3 = model_3.fit(train_images, train_labels, epochs=10, validation_split=0.15)  
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import itertools

# Evaluate the model's accuracy on the test dataset
test_loss_2, test_acc_2 = model_2.evaluate(test_images, test_labels)
print(f"Test Accuracy for Model 2: {test_acc_2}")

# Predicting the labels for the test set
predicted_labels_2 = model_2.predict(test_images)
predicted_labels_2 = np.argmax(predicted_labels_2, axis=1)

# Generating the confusion matrix
cm_2 = confusion_matrix(test_labels, predicted_labels_2)

# Plotting the confusion matrix
plt.figure(figsize=(10, 8))
plt.imshow(cm_2, interpolation='nearest', cmap=plt.cm.Blues)
plt.title('Confusion Matrix for Model 2')
plt.colorbar()
thresh = cm_2.max() / 2.
for i, j in itertools.product(range(cm_2.shape[0]), range(cm_2.shape[1])):
    plt.text(j, i, format(cm_2[i, j], 'd'),
             horizontalalignment="center",
             color="white" if cm_2[i, j] > thresh else "black")

tick_marks = np.arange(len(cm_2))
plt.xticks(tick_marks, range(len(cm_2)))
plt.yticks(tick_marks, range(len(cm_2)))
plt.ylabel('Actual Labels')
plt.xlabel('Predicted Labels')
plt.show()

# Identifying specific examples of incorrect classifications
incorrects_2 = np.nonzero(predicted_labels_2 != test_labels)[0]

#  Displaing the first 5 incorrect classifications
for i, incorrect in enumerate(incorrects_2[:5]):
    plt.imshow(test_images[incorrect], cmap='gray', interpolation='none')
    plt.title(f"True: {test_labels[incorrect]}, Predicted: {predicted_labels_2[incorrect]}")
    plt.show()
```
