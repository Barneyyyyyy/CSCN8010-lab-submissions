```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 2, 5)  # Sample data.

plt.figure(figsize=(5, 10), layout='constrained')
plt.plot(x, x, label='buses in town')  # Plot some data on the (implicit) axes.
plt.plot(x, x**5, label='cars in town')  # etc.
plt.plot(x, x**3, label='cycles in town')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Simple Plot")
plt.legend()
```

```python
import plotly.express as px
df = px.data.medals_long()

fig = px.area(df, x="medal", y="count", color="nation",
             pattern_shape="nation", pattern_shape_sequence=[".", "x", "/"])
fig.show()
```

```python
x = np.arange(5)
y1 = [10, 30, 60, 70, 100]
y2 = [20, 40, 50, 80, 90]
width = 0.40
  
# plot data in grouped manner of bar type
plt.bar(x-0.2, y1, width)
plt.bar(x+0.2, y2, width)
```

```python
fig, ax = plt.subplots(figsize=(5, 2.7))

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2 * np.pi * t)
line, = ax.plot(t, s, lw=2)

ax.annotate('max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))

ax.set_ylim(-2, 2)
```
