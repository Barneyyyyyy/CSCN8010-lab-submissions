```python
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt

```

# Pie Chart

**Display the vehicle sold details in 2022**

```python
labels = 'bmw','audi','porsche','maruti'
sizes = [15, 30, 45, 10]
fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['olivedrab', 'rosybrown', 'gray', 'saddlebrown'])
ax.set_title('vehicle sale in 2022',
             fontweight ="bold")
```

# Grouped bar chart with labels

**Display the grouped barchart**

```python
x = np.arange(10)
y1 = [10, 30, 60, 70, 100,20,25,45,50,30]
y2 = [20, 40, 50, 80, 90, 65, 75, 85, 25, 80]
bar_colors = ['tab:red', 'tab:blue']
width = .20 #set the width of the bar
  
# plot data
plt.bar(x-0.2, y1, width)
plt.bar(x+0, y2, width)
```

# Bar Chart

**Control the bar color with label parameter**

```python
fig, ax = plt.subplots(figsize=(4,5))

fruits = ['tomatto', 'chilly', 'brinjal', 'onion']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'green', 'black', 'purple']
bar_colors = ['tab:red', 'tab:blue', 'black', 'purple']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('vegitables availability',fontsize=12,fontweight='bold')
ax.legend(title='vegitable color')

plt.show()
```

# Annotating a plot

 **how to annotate a plot with an arrow pointing in a chart**

```python
fig, ax = plt.subplots(figsize=(7,3))

t = np.arange(0.0, 5.0,0.5)
s = np.cos(2*np.pi*t)
line, = ax.plot(t, s, lw=2)

ax.annotate('the local max', xy=(4, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='red'),
            )
ax.set_ylim(-2, 2)
plt.show()
```
