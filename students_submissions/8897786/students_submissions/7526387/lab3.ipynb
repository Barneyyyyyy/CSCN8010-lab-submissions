### Step 1 - Frame the problem and look at the big picture.

>This lab seeks to explore the relation between BMI and the measure of disease progression one year after baseline

### Step 2 - Get the data

```python
import numpy as np
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt  
from sklearn import datasets
# diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True, scaled=False, as_frame=True)
data_X, data_y = datasets.load_diabetes(return_X_y=True, scaled=True, as_frame=False)
data_X = data_X[:, 2].reshape((-1,1))
```

### Step 3 - Explore the data to gain insights

```python
df_X, df_y = datasets.load_diabetes(return_X_y=True, scaled=False, as_frame=True)
```

```python
df_X
```

The spread of the data is quite large.

```python
plt.scatter(df_X['bmi'], df_y,  color='red')
plt.show()
```

### Step 6 - Fine-tune your models and combine them into a great solution

```python
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y, test_size=0.2, random_state=0)
model = LinearRegression()  
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

### Step 7 - Present your solution

```python
print('Mean squared error: %.2f' % metrics.mean_squared_error(y_test, y_pred))
print('Intercept: %.2f' % model.intercept_)
print('Slope: %.2f' % model.coef_)
```

The Fit of the graph seems relitively good regardless of the high error value. I belive the high MSE is due to the large spread of values given. It is know that higher BMI leads to health complications and this is reflected by the model.

```python
plt.scatter(X_test, y_test,  color='red')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.show()
```
