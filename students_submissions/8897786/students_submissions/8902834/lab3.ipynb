<h3><B><U>Importing the necessary packages:</B></U></h3>
<h5>We need to import the packages that are necessary for us to find and plot linear regression. Numpy, matplotlib, and sklearn are some of the packages that we will be requiring.</h5>

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

<h3><B><U>Loading the Diabetes Dataset:</U></B></h3>
<h5>To find the linear regression the most important part is to get related data. The larger the data, more efficient is the model. We will now have to load or import the dataset so we can create a model for it. </h5>

```python
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

<h3><u><b>Splitting the data:</b></u></h3>
<h5>In univarient linear regression we work with only one indepent attribute. In this case we are using the BMI attribute and so we need to only extract the BMI column from the dataset. We split the dataset into two parts. One is the training dataset that has more data and the other is testing data which has less number of data. The training dataset is larger than testing because we need  to train our model to be more efficient. Testing dataset is used to validate if are model is working correctly.</h5>

```python

diabetes_bmi = diabetes_X[:,2].reshape((-1,1))
X_train,X_test,y_train,y_test = train_test_split(diabetes_bmi, diabetes_y, test_size=0.2, random_state= 0)
```

<h3><U><B>We built a Linear Regression Model that fits our Training Datasets:</B></U></h3>
<h5>

```python
model = LinearRegression()
model.fit(X_train,y_train)
```

<h3><B><U>The Model's Parameter - Slope(m) and Intercept(c):</U></B></h3>

```python
m = float(model.coef_)
c = model.intercept_
y_pred = model.predict(X_test)
print("The slope of the model is m =",m)
print("The intercept of the model is c =",c)
```

<h3><u><b>The loss on the Training and Testing Datasets:</B></U></h3>

<h5>The loss in the training or testing dataset is a measure of how well the model fits the training or testing data. It quantifies the discrepancy between the actual target values (X_train)(y_train) and the predicted values generated by the model. The lower the loss, the better the model is at fitting the training or testing data. However, it is important to note that a low training or testing loss does not guarantee good performance on unseen data. It is essential to evaluate the model's performance on a separate test dataset to assess its generalization capability.</h5>

```python
y_train_pred = model.predict(X_train)
train_loss = np.mean((y_train - y_train_pred) ** 2)
test_loss = np.mean((y_test - y_pred) ** 2)
print("The loss on the training dataset is", train_loss)
print("The loss on the testing dataset is", test_loss)
```

<h3><u><b>A graph representing the Training Dataset:</h3></u></b>

```python
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Training Data')
plt.legend()
plt.show()

```

<h3><u><b>A graph representing the Testing Datatset:</h3></u></b>

```python
plt.scatter(X_test, y_test, color='blue', label='Testing Data')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Testing Data')
plt.legend()
plt.show()
```

<h3><u><b>A graph respresenting the Linear Regression - Model:</h3></u></b>

```python
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Model')
plt.legend()
plt.show()
```

<h3><u><b>A graph respresenting the Linear Regression - Model, Training and Testing Dataset:</h3></u></b>

```python


# Plot the test data and the linear regression model
plt.scatter(X_train, y_train, color='magenta', label='Training Data')
plt.scatter(X_test, y_test, color='blue', label='Test Data')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Linear Regression Model')
plt.xlabel('BMI')
plt.ylabel('Disease Progression')
plt.title('Linear Regression - Model, Training and Testing')
plt.legend()
plt.show()
```

<h3><u><b>Takeaways:</h3></u></b>

<h5>1. The model's slope (m) indicates the rate of change in disease progression per unit change in BMI. A positive value suggests that higher BMI is associated with increased disease progression.<br/>
2. The intercept (c) represents the estimated disease progression when BMI is zero. However, in the context of BMI, an intercept of zero is not meaningful since BMI cannot be zero in reality.<br/>
3. The loss on the test dataset provides an indication of how well the model generalizes to unseen data. A lower loss value indicates better performance but we have a high loss in training and testing dataset.<br/><br/>

To conclude, the univariate linear regression model on the diabetes dataset provides a reasonable fit, suggesting that BMI may be a useful predictor for disease progression. However, further analysis and evaluation of additional features could provide a more comprehensive understanding of the factors influencing disease progression in diabetes. Using the remaining attributes we can train our model much better and get good results. </h5>
