```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

# Format string
we can use shortcut string notation parameter to specify the marker

```python
y = np.array([5, 14, 4, 12])
plt.plot(y, 's--g') # format string marker|line|color :
                    # s => square
                    # -- => dashed line 
                    # g => green color
plt.show()
```

# Piechart
We can use pie() function to draw pie chart. 
We can use the legend() function to add a list of explanation for each wedge. 

```python
count = np.array([1407563.84,38246.11, 1412360.00,125681.59]) # X-axis
countries = ["India", "Canada", "China", "Japan"] # Y-axis

plt.pie(count, labels = countries)
plt.legend(title = "Population")  # add header to the legend
plt.show() 
```

# Grid 
We can use grid() function to add grid lines to the plot

```python
time = np.array([8, 10, 12, 14, 16, 18, 20]) #x- axis
temp = np.array([8, 27, 32, 17, 14,12,4]) # y-axis

plt.title("Weather")
plt.xlabel("Time")
plt.ylabel("Temperature")

plt.plot(time, temp)

plt.grid() # Add grid view

plt.show()
```

# Horizontal Bar graph
barh() function draws the horizontal graph

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.title("Score Board")
plt.xlabel("Students")
plt.ylabel("Grade")
plt.barh(x, y, color='Purple') # Plot horizontal plot
plt.show()
```
