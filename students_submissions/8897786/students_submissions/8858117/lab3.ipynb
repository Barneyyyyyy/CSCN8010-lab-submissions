# Practical Lab 3 - Univariate Linear Regression

Importing the required packages

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
```

Importing the diabetes dataset

```python
from sklearn import datasets
diabetes = datasets.load_diabetes()
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
```

Dataset Details

```python
print(diabetes.DESCR)
```

```python
diabetes.feature_names
```

```python
diabetes_X.shape
```

```python
diabetes_y.shape
```

Defining feature and split data into training and testing sets

```python
diabetes_X= diabetes_X[:, np.newaxis,2]
x_train, x_test, y_train, y_test = train_test_split(diabetes_X, diabetes_y, test_size=0.2, random_state=0)
```

Fitting Simple Linear Regression to the Training set

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

 Predicting the Test set results

```python
y_pred = model.predict(x_test)
```

```python
print('Mean squared error: %.2f' % metrics.mean_squared_error(y_test, y_pred))
```

Visualising the Test set results

```python
plt.scatter(x_test, y_test, color='green')
```

Visualizing the prediction

```python
plt.plot(x_test, y_pred, color='orange')
```

Visualizing the data distribution

```python
plt.scatter(x_train, y_train, color='green')
```

Visualizing the model to get an insight of the fitness of the model

```python
plt.scatter(x_test, y_test, color='purple')
plt.plot(x_test, y_pred, color='green')
plt.xlabel("Feature 3: BMI")
plt.ylabel("Target: Diabetes progression one year after baseline")
plt.title("Training: Linear Regression between bmi and diabetes")
plt.show()
```

Evaluate the Model

```python
# slope
m = model.coef_[0]
print('Coefficients/Slope: ', m)

# intercept
b = model.intercept_
print('Intercept: ', b)

# Mean Abosolute Error
mae = mean_absolute_error(y_test, y_pred)
print('Mean Absolute Error: ', mae)

# Mean Squared Error
mse = mean_squared_error(y_test, y_pred)
print('Mean Squared Error: ', mse)

# Root Mean Square Error
print("RMSE of the model is: %.2f" %np.sqrt(mse))
```


This univariate regression model has a higher mean square error and mean absolute. There is a significant deviation in the prediction results from the actual values.
So, we can not consider this a good prediction model.
