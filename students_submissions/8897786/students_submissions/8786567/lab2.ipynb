```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot
A diagram which displays the relationship between two numerical variables.

```python
np.random.seed(58369)  # seed the random number generator.
data = {'a': np.arange(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b', data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

## Plot
Draws a line from point to point.

```python
x = np.linspace(-np.pi,np.pi,100)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x, np.sin(x), label='Trigonometric')
```

## Pie Chart
A circular plot which can only display one series of data.

```python
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']
 

data = np.random.rand(6)
 
# Creating plot
fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = cars)
 
# show plot
plt.show()
```

## Bar Graph
A graphical display of categorical data using bars of different heights.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
langs = ['English', 'French', 'Spanish', 'Chinese']
students = np.arange(4)
ax.bar(langs,students)
plt.show()
```
