```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot
The numpy scatter function will plot a dot for each data entry and requires two arrays of equal length. One array for the x values and the other for the y values.

```python
np.random.seed(123)  # seed the random number generator.
data = {'a': np.arange(50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.scatter('a', 'b',  data=data)
ax.set_xlabel('entry a')
ax.set_ylabel('entry b')
```

## Plot
The plot function draws a line through the given data points.

```python
x = np.linspace(0, 2, 100)

fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
ax.plot(x, x**2, label='quadratic')
ax.set_xlabel('x label')
ax.set_ylabel('y label')
```

## Histogram
A histogram is a graphical representation of the frequency distribution of data.

```python
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10001)
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')

# the histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='C0', alpha=0.75)

ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Aardvark lengths\n (not really)')
ax.text(75, .025, r'$\mu=115,\ \sigma=15$')
ax.axis([55, 175, 0, 0.03])
ax.grid(True)
```

## Bar
The bar graph represents categorical data with rectangular bars with heights that are proportional to the values.

```python
fig, ax = plt.subplots(figsize=(5, 2.7), layout='constrained')
categories = ['turnips', 'rutabaga', 'cucumber', 'pumpkins']

ax.bar(categories, np.random.rand(len(categories)))
```
