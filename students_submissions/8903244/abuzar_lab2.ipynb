Line Plot
This code generates a simple line plot of the sine function.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine Function')
plt.show()
```

bar plot
This code creates a bar plot showing the values for different categories.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 50, 75, 100]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
plt.show()
```

Scatter plot
This code generates a scatter plot with randomly generated data points. 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)

plt.scatter(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot')
plt.show()
```

Histogram 
This code creates a histogram of a random dataset.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
data = np.random.normal(0, 1, 1000)

plt.hist(data, bins=20)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```
