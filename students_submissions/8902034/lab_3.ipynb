```python

from sklearn import datasets


# Load the diabetes dataset
diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y = True)

print("Feature matrix shape:", diabetes_X.shape)
print("Target vector shape:", diabetes_y.shape)
feature_names = datasets.load_diabetes().feature_names
print("Feature names:", feature_names)

import numpy as np

# Calculate mean, standard deviation, minimum, and maximum of each feature
feature_stats = {
    "Mean": np.mean(diabetes_X, axis=0),
    "Standard Deviation": np.std(diabetes_X, axis=0),
    "Minimum": np.min(diabetes_X, axis=0),
    "Maximum": np.max(diabetes_X, axis=0)
}
for feature_name, stats in feature_stats.items():
    print(f"{feature_name}:\n{stats}\n")
    import matplotlib.pyplot as plt

# Plotting a histogram of the target variable
plt.hist(diabetes_y, bins=20)
plt.xlabel("Disease Progression")
plt.ylabel("Count")
plt.title("Histogram of Disease Progression")
plt.show()

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(diabetes_X, diabetes_y, test_size=0.2, random_state=42)

# Create an instance of Linear Regression
linear_regression = LinearRegression()

# Fit the model to the training data
linear_regression.fit(X_train, y_train)

# Make predictions on the testing data
y_pred = linear_regression.predict(X_test)

# Evaluate the model using Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Linear Regression MSE:", mse)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Fit the model to the entire dataset
linear_regression.fit(diabetes_X, diabetes_y)

# Make predictions on the entire dataset
y_pred = linear_regression.predict(diabetes_X)

# Plot the scatter plot with regression line
plt.scatter(diabetes_X[:, 2], diabetes_y, color='blue', label='Actual')
plt.plot(diabetes_X[:, 2], y_pred, color='red', linewidth=2, label='Linear Regression')
plt.xlabel('Body Mass Index')
plt.ylabel('Disease Progression')
plt.title('Linear Regression Scatter Plot')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Calculate the mean disease progression for each feature
feature_names = datasets.load_diabetes().feature_names
mean_progression = np.mean(diabetes_X, axis=0)

# Plot the bar graph
plt.figure(figsize=(8, 6))
plt.bar(feature_names, mean_progression)
plt.xlabel('Features')
plt.ylabel('Mean Disease Progression')
plt.title('Mean Disease Progression by Feature')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
