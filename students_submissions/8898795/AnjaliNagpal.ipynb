```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 1. Pie chart using Matplotlib

```python
x=[4,9,16,25,36]
fig = plt.figure(figsize =(4.5, 4.5))

plt.pie(x, labels=("Guavas", "Berries","Mangoes","Apples", "Avocado"),
colors = ( "mistyrose", "#fA00ff", "#ebc713", "#ff0000", "#8aeb13"))

plt.show()
```

**Fruits grown in a region**
|   Fruits  |    Percentage grown    |
|   :----:  |   :----:  |
|   Guavas  |    4%     |
|   Berries |   9%  |
|Mangoes|16%|
|Apples|25%|
|Avocado|36%|


### 2. Tan(x) Graph

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 5, 100)
plt.plot(x, np.tan(x), linewidth=4)
plt.grid(color='black', alpha=0.3, linestyle='--', linewidth=2)
plt.show()
```

### 3. Multiline Graph

```python
x = [1,2,3,4,5]
y = [3,3,3,3,3]

plt.plot(x, y, label = "Horizontal line", linestyle="-")
plt.plot(y, x, label = "Vertical line", linestyle="--")
plt.plot(x, np.sin(x), label = "Sin(x)", linestyle="-.")
plt.plot(x, np.cos(x), label = "Cos(x)", linestyle=":")
plt.legend()
plt.show()
```

```python
df=pd.DataFrame({'x_values': range(1,11), 'y1_values': np.random.randn(10), 'y2_values': np.random.randn(10)+range(1,11), 'y3_values': np.random.randn(10)+range(11,21) })
plt.plot( 'x_values', 'y1_values', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'x_values', 'y2_values', data=df, marker='X', color='orange', linewidth=2)
plt.plot( 'x_values', 'y3_values', data=df, marker='.', color='violet', linewidth=2, linestyle='dashed', label="dash")

plt.legend()
plt.show()
```

### 4. Histogram

```python
df = pd.DataFrame({
    'Name': ['Chritine', 'Beth', 'Flora','Christopher'],
    'Age': [90, 30, 45, 15],
    'Height(in cm)': [135,150, 180, 160]
})
ax = df.plot(x="Name", y="Height(in cm)", kind="bar", color="cyan")
df.plot(x="Name", y="Age", kind="bar", ax=ax, color="royalblue")
```
